```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import StratifiedKFold
from xgboost import XGBClassifier
```

```python
df_listo = pd.read_csv('df_listo.csv')
df_nuevo = df_listo.loc[df_listo['Nombre de la entidad_codificado'] != 24]
df_nuevo
```

```python
y = df_listo['Nombre de la entidad_codificado'].values 
X = df_listo.drop(['Nombre de la entidad_codificado'], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
params = {
          "learning_rate" : [0.05,0.10,0.15,0.20,0.25,0.30],
          "max_depth" : [ 3, 4, 5, 6, 8, 10, 12, 15],
          "min_child_weight" : [ 1, 3, 5, 7 ],
          "gamma": [ 0.0, 0.1, 0.2 , 0.3, 0.4 ],
          "colsample_bytree" : [ 0.3, 0.4, 0.5 , 0.7 ]
        }
```

```python
xgb = XGBClassifier(learning_rate=0.02, n_estimators=600, objective='binary:logistic', silent=True, nthread=1)
folds = 3
param_comb = 5
xgb_RandomGrid = RandomizedSearchCV(estimator=xgb, param_distributions=params, n_iter=param_comb, scoring='accuracy', n_jobs=4, cv=3, verbose=3, random_state=1001 )
xgb_RandomGrid.fit(X_train, y_train)
print (xgb_RandomGrid.best_params_)
print("Best score: ", xgb_RandomGrid.best_score_)
```
