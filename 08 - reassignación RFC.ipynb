```python
import numpy as np
import pandas as pd
```

```python
df_listo = pd.read_csv('df_listo.csv')
df_listo
```

```python
beneficiario_SURA = df_listo.loc[df_listo['Nombre de la entidad_codificado'] == 24]
print("Las entradas para los beneficiarios SURA son:")
print(beneficiario_SURA)
```

```python
df_nuevo = df_listo.loc[df_listo['Nombre de la entidad_codificado'] != 24]
df_nuevo
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,classification_report
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn.multiclass import OneVsRestClassifier 
from sklearn.model_selection import RandomizedSearchCV
```

```python
y = df_nuevo['Nombre de la entidad_codificado'] 
X = df_nuevo.drop(['Nombre de la entidad_codificado'], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
rfc =RandomForestClassifier(n_estimators = 80, min_samples_split = 5, min_samples_leaf = 1, max_features = 'log2', max_depth = 4, bootstrap = False)
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
from sklearn.metrics import accuracy_score
print((accuracy_score(y_test, y_pred)))
```

```python
def predecir_entidad(datos_paciente):

    df_paciente = pd.DataFrame([datos_paciente])
    
    prediccion = rfc.predict(df_paciente)
    
    return prediccion[0]
```

```python
#Tomando la entrada 9, primer perteneciente SURA

datos_nuevo_paciente = {
    'Tipo de afiliado_ADICIONAL': 0, 
    'Tipo de afiliado_BENEFICIARIO': 1, 
    'Tipo de afiliado_COTIZANTE': 0, 
    'Estado del afiliado_Activo': 1, 
    'Estado del afiliado_Protección Laboral C':0, 
    'Condición del beneficiario_DISCAPACIDAD': 0, 
    'Condición del beneficiario_ESTUDIANTE': 0, 
    'Condición del beneficiario_NO APLICA': 1, 
    'Zona de Afiliación_Rural': 0, 
    'Zona de Afiliación_Urbana': 1,
    'Género_Femenino': 1,
    'Género_Masculino': 0,
    'Grupo etario_codificado': 0,  
    'Departamento_codificado':30
}

entidad_predicha = predecir_entidad(datos_nuevo_paciente)
print(f'La entidad predicha para el nuevo paciente es: {entidad_predicha}')
```
